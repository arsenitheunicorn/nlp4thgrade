```python
import pandas as pd
from nltk.tokenize import sent_tokenize
import spacy
import re
```

```python
import gzip
import json

def parse(filename):
    f = gzip.open(filename, 'rt')
    entry = {}
    for l in f:
        l = l.strip()
        colonPos = l.find(':')
        if colonPos == -1:
            yield entry
            entry = {}
            continue
        eName = l[:colonPos]
        rest = l[colonPos+2:]
        entry[eName] = rest
    yield entry

raw_data = []
    
for e in parse("Musical_instruments.txt.gz"):
    raw_data.append(json.loads(json.dumps(e)))
    
df = pd.json_normalize(raw_data)
df
```

Опираться на одни названия тут достаточно ненадежно, причем по нескольким причинам:

 - во-первых, множество музыкальных аксессуаров (струны, кабели, чехли, ремни, мембраны и проч.) содержат в своих названиях названия инструментов; если и использовать заголовки, то только с каким-либо иерархическим поиском;
 - названия инструментов могут содержаться в книгах или фильмах; а могут не содержаться нигде
 
Работа с заголовками потребует от нас искусственно ограничить список товарных категорий, которые мы будем рассматривать в этом анализе.

Также можно включать несколько категорий в один товар – все же есть разница между, допустим, струнами на бас-гитару и на мандолину.

По сэмплу текстов отзывов возникает ощущение, что те категории, которые можно выделить из названия, достаточно встречаются в __первом__ предложении отзыва, это тоже можно будет учитывать.

Список категорий из файла, представленного в следующем cell'е, составлен с опорой на [википедию](https://simple.wikipedia.org/wiki/List_of_musical_instruments) и личный опыт.

```python
nlp = spacy.load('en_core_web_sm')
```

```python
with open('catlist.txt', 'r', encoding='utf-8') as f:
    CATEGORIES = f.read().split('\n')
```

```python
def amp(L):
    # this is too often of a contraction
    if 'amp' in L:
        if 'amplifier' not in L:
            L.append('amplifier')
        L.remove('amp')
    return L
    
    
def extract_categ_title(title, C=CATEGORIES):
    if re.search('\(\d{4}\)', title):
        return []
    
    title = title.lower()
    output = []
    lemmas = [x.lemma_ for x in nlp(title)]
    for T in C:
        if T in lemmas:
            output.append(T) 
    
    return amp(output)

def extract_first_sent(text, C=CATEGORIES):
    text = nlp(sent_tokenize(text)[0].lower())
    output = []
    lemmas = [x.lemma_ for x in text]
    for T in C:
        if T in lemmas:
            output.append(T) 
    
    return amp(output)

def categories(L1, L2):
    return list(set(L1) | set(L2))


features = []
for i, row in df.iterrows():
    L1 = extract_categ_title(row['product/title'])
    L2 = extract_first_sent(row['review/text'])
    features.append(categories(L1, L2))
    if i % 2000 == 0:
        print(i)
    
print(len(features))
```

Теперь будем смотреть на всякие коллокации. Сразу приходят на ум следующие варианты (word - одно из слов категории):

- ADJ + word (cool guitar)
- 'this'? + word + VERB + ADJ (this bass sounds great)
- %verb_of_feelings% + ? + word



```python
from nltk.corpus import wordnet


def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from spacy.matcher import Matcher

```

```python
FEELINGS = ['love', 'hate', 'like', 'admire']
DET = ['this', 'that', 'its']

def NE_groups(text, features, 
              feelings=FEELINGS, DET=DET):
    
    matcher = Matcher(nlp.vocab)

    if 'fog machine' in features:
        features = features.remove('fog machine')
    NE = features + ['tool', 'instrument', 'it', 'sound']
    
    patterns = [
        [{'POS': 'ADJ'}, {'LEMMA': {'IN': NE}}],
        [{'LEMMA': {'IN': NE}}, {'POS': {'IN': ['AUX', 'VERB']}}, {'POS': 'ADJ'}],
        [{'LEMMA': {'IN': feelings}}, {'LEMMA': {'IN': NE}}],
        [{'LEMMA': {'IN': feelings}}, {'LEMMA': {'IN': DET}}, {'LEMMA': {'IN': NE}}]
    ]
    
    matcher.add('ADJ_W', None, patterns[0])
    matcher.add('W_V_ADJ', None, patterns[1])
    matcher.add('feel_W', None, patterns[2])
    matcher.add('feel_DET_W', None, patterns[3])
    
    d = nlp(text.lower())
    found = matcher(d)
    NE_groups = [d[start:end].text for _, start, end in found]
    return NE_groups
```

```python
NE_gs = []

for i, row in df_feat.iterrows():
    if i % 2000 == 0:
        print(i)
    NE_gs.append(NE_groups(row['review/text'], row['features']))
    
df_feat['NE'] = NE_gs
df_feat.sample(20)
```

```python
from nltk.collocations import BigramAssocMeasures, BigramCollocationFinder

bam = BigramAssocMeasures()

NE_bcf = []
to_be = ['is', 'are', 'am', 'was', 'were']

for neg in NE_gs:
    for negL in neg:
        ng = negL.split(' ')
        for stopword in to_be:
            if stopword in ng:
                ng.remove(stopword)
        NE_bcf.append(ng)

bcf = BigramCollocationFinder.from_documents(NE_bcf)

bcf.score_ngrams(bam.pmi)[:30]
```

```python
bcf.score_ngrams(bam.student_t)[:30]
```

```python
bcf.score_ngrams(bam.likelihood_ratio)[:30]
```

```python
bcf.score_ngrams(bam.jaccard)[:30]
```

PMI и Jaccard выдают достаточно странные и в основном негативные результаты; t-score и likelihood справляются на первый взгляд гораздо лучше

```python
pmi = bcf.score_ngrams(bam.pmi)
jac = bcf.score_ngrams(bam.jaccard)
lklhd = bcf.score_ngrams(bam.likelihood_ratio)
t_score = bcf.score_ngrams(bam.student_t)
```

```python
def find_category(cat, measure, n=0):
    output = []
    for pair, score in measure:
        if cat in pair:
            output.append((pair, score))
    if n == 0:
        return output
    else:
        return output[:n]
```
