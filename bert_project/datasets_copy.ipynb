```python
big_dataset = pd.read_csv("tcc_ceds_music.csv")
```

```python
very_big_dataset = big_dataset[["artist_name", "lyrics", "release_date"]]
```

```python
#df = big_dataset[big_dataset['release_date'] > 1994][["artist_name", "lyrics"]]
```

```python
gender_df = pd.read_csv('singers_gender.csv')
female_artists = [x.lower() for x in gender_df[gender_df['gender'] == 'female']['artist'].to_list()]
male_artists = [x.lower() for x in gender_df[gender_df['gender'] == 'male']['artist'].to_list()]
len(female_artists), len(male_artists)
```

```python
df_selected_artists = pd.read_excel('eng_artists.xlsx', header=0)
#male_artists = df_selected_artists['men'].to_list()
female_artists.extend(df_selected_artists['women'].to_list())
female_artists = list(set(female_artists))
len(female_artists)
```

```python
male_artists.extend(df_selected_artists['men'].to_list())
male_artists = list(set(male_artists))
```

```python
len(male_artists)
```

```python
male_df = very_big_dataset[very_big_dataset["artist_name"].isin(male_artists)]
fem_df = very_big_dataset[very_big_dataset["artist_name"].isin(female_artists)]
```

```python
male_df['gender'] = 1
fem_df['gender'] = 0
```

```python
len(male_df.index), len(fem_df.index)
```

```python
sample_male_all = male_df.sample(2913)
sample_male_all
```

```python
def gendering(x, male_artists=male_artists, female_artists=female_artists):
    if x in male_artists:
        val = 1
    elif x in female_artists:
        val = 0
    else:
        val = 2
    return val
```

```python
tc_sample_male = sample_male_all[['lyrics', 'gender']]
tc_fem_df = fem_df[['lyrics', 'gender']]
len(tc_sample_male), len(tc_fem_df)
```

```python
df_all = pd.concat([tc_sample_male, tc_fem_df])
df_all.groupby("gender").count()
```
